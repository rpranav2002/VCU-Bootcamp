# Assignment 1: Data Audit and Exploratory Data Analysis (EDA)

## Dataset: Online Retail

This notebook performs a data audit and exploratory data analysis (EDA) on the Online Retail dataset.


```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_excel('Online Retail.xlsx')

# Display first few rows
df.head()
```

## Dataset Overview

```
# Dataset shape and column information
print("Shape:", df.shape)
print("\nColumns:")
print(df.columns.tolist())

df.info()
```

## Missing Values

```
# Missing value analysis
missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing Values': missing_values,
    'Percentage': missing_percentage
})

missing_df
```

## Duplicate Records

```
# Check for duplicates
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")
```

## Summary Statistics

```
# Descriptive statistics for numerical variables
df[['Quantity', 'UnitPrice']].describe()
```

## Distribution Analysis

```
# Distribution of Quantity
plt.figure(figsize=(8,5))
df['Quantity'].clip(-100, 100).hist(bins=50)
plt.title('Distribution of Quantity')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.show()

# Distribution of UnitPrice
plt.figure(figsize=(8,5))
df['UnitPrice'].clip(0, 100).hist(bins=50)
plt.title('Distribution of Unit Price')
plt.xlabel('Unit Price')
plt.ylabel('Frequency')
plt.show()
```

## Outlier Detection

```
# Boxplots for outlier detection
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
sns.boxplot(x=df['Quantity'])
plt.title('Quantity Boxplot')

plt.subplot(1,2,2)
sns.boxplot(x=df['UnitPrice'])
plt.title('Unit Price Boxplot')

plt.tight_layout()
plt.show()
```

## Relationship Exploration

```
# Scatter plot of Quantity vs UnitPrice
sample_df = df.sample(5000, random_state=42)

plt.figure(figsize=(8,6))
sns.scatterplot(data=sample_df, x='Quantity', y='UnitPrice', alpha=0.5)
plt.title('Quantity vs Unit Price')
plt.xlabel('Quantity')
plt.ylabel('Unit Price')
plt.show()
```

## Country Analysis

```
# Top countries by transaction count
top_countries = df['Country'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title('Top 10 Countries by Number of Transactions')
plt.xlabel('Transaction Count')
plt.ylabel('Country')
plt.show()
```

## Key Findings

1. The dataset contains over 541,000 transaction records and 8 variables.
2. CustomerID has a substantial number of missing values (~25%).
3. Approximately 5,268 duplicate rows were identified.
4. Quantity and UnitPrice exhibit significant skewness and contain extreme outliers.
5. The United Kingdom accounts for the majority of transactions.
6. Negative quantities and prices likely represent returns, cancellations, or data entry issues.
